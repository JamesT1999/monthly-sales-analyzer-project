Calcula las ventas totales de un producto dado (por ejemplo, "product_a") a lo largo de 20 días.

```python
def total_sales_by_product(data, product_key):
    """Calculates the total sales of a specific product in 30 days."""
    total = 0
    
    for day in data:
        total += day.get(product_key, 0)
    
    return total
```

```python
print(total_sales_by_product(sales_data, "product_a"))
```

Calcula el promedio de ventas diarias de un producto dado.

```python
def average_daily_sales(data, product_key):
    """Calculates the average daily sales of a specific product."""
    if not data:
        return 0
    
    total = 0
    
    for day in data:
        total += day.get(product_key, 0)
    
    return total / len(data)

```

```python
print("Average daily sales of product_b:", average_daily_sales(sales_data, "product_b"))
```

Encuentra el día con las ventas totales más altas (suma de los tres productos).

```python
def best_selling_day(data):
    """Finds the day with the highest total sales."""
    best_day = None
    max_sales = 0

    for day in data:
        total = day.get("product_a", 0) + day.get("product_b", 0) + day.get("product_c", 0)
        
        if total > max_sales:
            max_sales = total
            best_day = day["day"]

    return best_day, max_sales
```

```python
print("Day with highest total sales:", best_selling_day(sales_data))
```

Cuenta cuántos días las ventas de un producto superaron un umbral dado. (por ejemplo, 18).

```python

def days_above_threshold(data, product_key, threshold):
    """Counts how many days the sales of a product exceeded a given threshold."""
    count = 0
    
    for day in data:
        if day.get(product_key, 0) > threshold:
            count += 1
    
    return count
```

```python
print("Days when product_c exceeded 300 sales:", days_above_threshold(sales_data, "product_c", 300))
```

 Identifica qué producto (A, B o C) tuvo las ventas totales más altas.

```python
def top_product(data):
    """Determines which product had the highest total sales in 30 days."""
    totals = {
        "product_a": 0,
        "product_b": 0,
        "product_c": 0
    }

    for day in data:
        for product in totals:
            totals[product] += day.get(product, 0)

    top = max(totals, key=totals.get)
    
    return top, totals[top]
```

```python
print("Product with highest total sales:", top_product(sales_data))
```
